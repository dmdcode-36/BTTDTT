```python
#W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không?
def snt(n):
    if n==2:
        return True
    elif n>2:
        for i in range (2,n):
            if n%i==0:
                return False
            else:
                return True
    return False
n=int(input())
print(snt(n))
```

```python
#W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên.
#Số hoàn hảo (hay số hoàn chỉnh) là số nguyên dương có tổng các ước nguyên dương thực sự (không kể chính nó) bằng chính số đó.
#Ví dụ: 6 (\(1+2+3\)) và 28 (\(1+2+4+7+14\)) là các số hoàn hảo nhỏ nhất. Tất cả số hoàn hảo được tìm thấy đến nay đều là số chẵn, liên quan đến số nguyên tố Mersenne.
#Dạng số hoàn hảo chẵn: Được Euclid và Euler chứng minh có dạng \(2^{p-1} \times (2^p - 1)\), trong đó \(2^p - 1\) là một số nguyên tố Mersenne.
def is_perfect_number(n):
    if n <= 1:
        return False
    tong = 1  
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            tong+= i
            if i != n // i:
                tong+= n // i
    return tong == n
n=int(input())
print(is_perfect_number(n))    
```

```python
#W5A5 
#Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử đầu tiên có giá trị k trong list số nguyên, 
#nếu không có thì trả về -1 
def find_first_index(lst, k):
    for num in lst:
        if num==k:
            return lst.index(k)
    return -1
lst=list(map(int,input().split()))
k=int(input())
print(find_first_index(lst, k)) 
```

```python
#W5A7 Viết hàm giả lập một máy tính bỏ túi. 
#Hàm thực hiện tính toán kết quả của phép toán giữa số ( num1 ) và ( num2 ) với toán tử ( operat ) 
#Biết các toán tử đơn giản thực hiện phép toán lần lượt là: 
#Phép trừ ( - ) 
#Phép cộng ( + ) 
#Phép chia ( / ) 
#Phép nhân ( * ) 
#Kết quả làm tròn đến chữ số thập phân thứ hai. 
#Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)). 
#Đầu ra: Kết quả phép toán. 
a,b=map(float,input().split())
op=input()
def compute(a, b, op):
    if op == '/' and b == 0:
        return ValueError
    elif op !='+' and op !='-' and op !='*' and op !='/':
        return ValueError
    elif op =='+':
        return round(a+b,2)
    elif op =='-':
        return round(a-b,2)
    elif op =='*':
        return round(a*b,2)
    elif op =='/':
        return round(a/b,2)
compute(a, b, op)
```

```python
#W5A8 
#Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng của chúng.
#Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2. 
#Cho hai số nguyên  x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này. 
#Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý thuyết này trong tài liệu có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (error-detecting and error-correcting codes).
# Nó được sử dụng trong kỹ thuật viễn thông để tính số lượng các bit trong một từ nhị phân (binary word) bị đổi ngược, như một hình thức để ước tính số lỗi xảy ra trong quá trình truyền thông, và vì thế, đôi khi, nó còn được gọi là khoảng cách tín hiệu (signal distance). 
#Việc phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết tin học, lý thuyết mã hóa, và mật mã học. 
#Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác nhau, hay các dãy ký tự có xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn. 
def hamming_distance(x, y):
    kq = x ^ y
    kc = 0
    while kq > 0:
        kq &= (kq - 1)  
        kc += 1
    return kc
x,y=map(int,input().split())
print(hamming_distance(x,y))
```

```python
#W5A9 Viết hàm trả về tổng các chữ số của số nguyên dương n 
def sum_of_digits(n):
    total_sum = 0
    while n > 0:
        total_sum += n % 10 
        n //= 10           
    return total_sum
n=int(input())
print(sum_of_digits(n))  
```

```python
#W5A12 Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và 0 <= |r| < |divisor|.
def divide_with_remainder(dividend, divisor):
    q=dividend//divisor 
    r=dividend%divisor
    if dividend == divisor + r and 0 <= abs(r) < abs(divisor):
        return (q,r)
    return False
dividend,divisor=map(int,(input().split()))
divide_with_remainder(dividend, divisor)
```

```python
#W5A15 Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.
def is_right_triangle(a, b, c):
    if a+b>c and b+c>a and a+c>b :
        if a*a+b*b==c*c or a*a+c*c==b*b or b*b+c*c==a*a:
            return True
    return False
a,b,c=map(float,input().split())
is_right_triangle(a, b, c)
```

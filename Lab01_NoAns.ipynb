```python
#Bài 4 Viết hàm trả về tổng số số nguyên dương k sao cho 1 <= k < n và k > (n - 1) / 2. (Ví dụ: n=9 -> 4).
n=int(input())
def tsnd(n):
    a=0
    for k in range (n):
        if 1<=k<n and k>(n-1)/2:
            a+=1
    return a
print(f"n={n}->{tsnd(n)}")
```

```python
#Bài 6 Viết một hàm tính diện tích của một tam giác có độ dài các cạnh là a, b, c (sử dụng công thức Heron). Trả về lỗi ValueError nếu các cạnh không tạo thành một tam giác.
def dttg(a,b,c):
    if a + b > c and a + c > b and b + c > a:
        p = (a + b + c) / 2
        S = (p * (p - a) * (p - b) * (p - c))**0.5
        return S
    else:
        raise ValueError
try:
    a,b,c=map(float,(input().split()))
    print(dttg(a,b,c))
except ValueError:
    print('các cạnh không tạo thành một tam giác')
```

```python
#Bài 7 Hãy viết một hàm trả về True nếu a, b, c có thể là độ dài các cạnh của một tam giác; ngược lại trả về False.
def kttg(a,b,c):
    if a + b > c and a + c > b and b + c > a:
        return True
    return False
a,b,c=map(float,(input().split()))
print(kttg(a,b,c))
```

```python
#Bài 8 Hãy viết một hàm trả về ký tự đứng sau ký tự x ở vị trí thứ n trong bảng chữ cái.
#Giữ nguyên vỏ hộp và bọc xung quanh theo kiểu A..Z hoặc a..z.
def kt(n,x):
    if not n.isalpha():
        return n
    start = ord('A') if n.isupper() else ord('a')
    chumoi=(ord(n)-start+x)%26+start
    return (chr(chumoi))
n=input()
x=int(input())
print(kt(n,x))
```

```python
#Bài 9 Viết một hàm chuyển đổi chữ cái viết hoa thành chữ cái viết thường. Nếu đầu vào không phải là một ký tự đơn, hãy báo lỗi ValueError.
def t(n):
    if len(n)==1:
        a=n.lower()
    else:
        raise ValueError
    return a
try:
    n=input()
    print(t(n))
except ValueError:
    print("Không phải kí tự đơn")
```

```python
#Bài 10 Hãy viết một hàm, khi nhận vào một chữ cái thường, sẽ trả về một chuỗi chứa chữ cái đó lặp lại 10 lần.
def t(n):
    return n*10
n=input()
print(t(n))
```

```python
#Bài 12 Viết một hàm trả về một tam giác vuông cân có chiều cao n bằng cách sử dụng dấu '*'.
#Dòng i (được đánh số từ 1) có i '*', được nối với nhau bằng các dòng mới.
# Ví dụ n = 4
# *
# **
# ***
# ****
def vtg(n):
    tg=""
    for i in range (1,n+1):
        cot='*'*i
        tg+=cot+'\n'
    return tg
n=int(input())
print(vtg(n))
```

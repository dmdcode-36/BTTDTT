```python
#W5A16
#Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số nguyên
#Một số Armstrong (hay số tự luyến) là số bằng tổng các chữ số của nó mũ n, với n là số chữ số. Ví dụ: 153=1^3+5^3+3^3
def soa(n:int)->bool:
    nstr=str(n)
    mu=len(nstr)
    tong=sum(int(digit)**mu for digit in nstr )
    return tong==n
n=int(input())
print(soa(n))
```

```python
#W5A17
#Viết hàm kiểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x
def scp(n:int)->bool:
    a=int(n**0.5)
    return a*a==n
n=int(input())
print(scp(n))
```

```python
#W5A18 VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.
def dx(n: int) -> bool:
    if n < 0 or (n % 10 == 0 and n != 0):
        return False
    sdn = 0
    sbd = n
    while n> 0:
        so= n % 10
        sdn = sdn * 10 + so
        n //= 10
    return sbd == sdn
n=int(input())
print(dx(n))
```

```python
#W5A23 Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0. 
def quadratic_roots_count(a, b, c):
    if a == 0:
        if b == 0:
            if c==0:
                return "Luôn Đúng"
            else:
                return "Vô lý"
        else:
            return 1

    delta = b**2 - 4 * a * c

    if delta > 0:
        return 2
    elif delta == 0:
        return 1
    else:
        return 0
a,b,c=map(float,input().split())
print(quadratic_roots_count(a,b,c))
```

```python
#W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.
def are_collinear(x1, y1, x2, y2, x3, y3):
    area_determinant = x1 * (y2 - y3) + x2 * (y3 - y1) + x3 * (y1 - y2)
    return area_determinant == 0
x1, y1, x2, y2, x3, y3=map(float,input())
print(are_collinear(x1, y1, x2, y2, x3, y3))
```

```python
#W5A27 Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list) 
def sk(n):
    s=0
    for i in range(len(n)):
        if n[i]!=0:
            s+=1
    return s
n=list(map(int,input().split()))
sk(n)
```

```python
#W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).
import math
def gocabc(x1,y1,x2,y2,x3,y3):
    return math.degrees(math.acos((((x1-x2)*(x3-x2)+(y1-y2)*(y3-y2)))/((((x1-x2)**2+(y1-y2)**2)**0.5)*(((x3-x2)**2+(y3-y2)**2)**0.5))))
x1,y1,x2,y2,x3,y3=map(float,input().split())
gocabc(x1,y1,x2,y2,x3,y3)
```

```python
#W5A30 Viết hàm tính hệ số nhị thức C(n, k). 
import math
def tinh_he_so_nhi_thuc(n,k):
    if k < 0 or k > n:
        return 0
    return math.comb(n, k)
n,k=map(int,input().split())
tinh_he_so_nhi_thuc(n,k)
```

```python
#W5A34 Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau 
#Nếu n ≤ 0 trả về 0
def sodoimotkhacnhau(x: int) -> bool:
    s = str(x)
    return len(s) == len(set(s))

def sum_1_to_n(n: int) -> int:
    if n <= 0:
        return 0 
    tong = 0
    for x in range(1, n + 1):
        if sodoimotkhacnhau(x):
            tong += x
    return tong

n=int(input())
print(sum_1_to_n(n))
```

```python
#Bài 1: Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diện tích và chu vi của nó.
a,b=map(float,input().split())
print(f"Chu vi HCN là {(a+b)*2:.2f}",f"Diện tích HCN là {a*b:.2f}")
```

```python
#Bài 2: Nhập vào bán kính của hình tròn. In ra màn hình diện tích và chu vi của nó. 
r=float(input())
import math
print(f"Chu vi hình tròn là {math.pi*2*r:.2f}",f"Diện tích hình tròn là {math.pi*r**2:.2f}")
```

```python
#Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. 
#Kiểm tra xem nó có thể là độ dài ba cạnh hay không, 
#nếu  không thì in  ra màn  hình  'Khong la ba  canh cua mot tam  giac'. 
#Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường.
#Tính diện tích, chu vi của tam giác và in ra màn hình.
a,b,c=map(float,input().split())
p=(a+b+c)/2
if a+b>c and a+c>b and b+c>a:
    print (a,b,c,"Là độ dài ba cạnh tam giác")
else:
    print(a,b,c,"Không là ba cạnh của một tam giác")
if a==b!=c or a==c!=b or b==c!=a :
    print("Đây là tam giác cân")
elif a==b==c:
    print("Đây là tam giác đều")
elif a**2+b**2==c**2 or b**2+c**2==a**2 or a**2+c**2==b**2:
    print("Đây là tam giác vuông")
else:
    print("Đây là tam giác thường")
print(f"Chu vi tam giác là {p*2:.2f}",f"Diện tích tam giác là {(p*(p-a)*(p-b)*(p-c))**0.5:.2f}")
```

```python
#Bài 4: Viết chương trình giải phương trình bậc 1 (ax + b = 0)  
a,b=map(float,input().split())
if a==0:
    if b==0:
        print("PT vô số nghiệm")
    else:
        print("PT vô nghiệm")
elif a!=0 and b==0:
    print(f"PT có nghiệm duy nhất là x = {0}")
elif a!=0 and b!=0:
    print(f"PT có nghiệm duy nhất là x= {-b/a}")
```

```python
#Bài 10: Nhập vào tâm và bán kính của một đường tròn.
#Sau đó nhập vào một điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?
a,b,r,x,y=map(float,input().split())
if (x-a)**2+(y-b)**2==r**2:
    print(f"Điểm A({x},{y}) thuộc đường tròn")
else:
    print(f"Điểm A({x},{y}) không thuộc đường tròn")
```

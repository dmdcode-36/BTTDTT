```python
#W4A25 
#Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. 
#Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập
numbers = []
while True:
    n = int(input("Nhập một số nguyên (nhập -1 để dừng): "))
    if n == -1:
        break
    numbers.append(n)
if numbers:
    print(f"Số lớn nhất: {max(numbers)}")
    print(f"Số nhỏ nhất: {min(numbers)}")
else:
    print("Bạn chưa nhập số nào hợp lệ.")
```

```python
#W4A26 [Fibonaci1] 
#Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó 
#Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A.
A = int(input("Nhập số nguyên dương A: "))
f1, f2 = 1, 1
fib_max = 1
if A < 1:
    print("Không có số thỏa mãn.")
else:
    while f2 <= A:
        fib_max = f2  
        f1, f2 = f2, f1 + f2 
print(f"Số Fibonacci lớn nhất không vượt quá {A} là: {fib_max}")
```

```python
#W4A27 
#Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng) 
#Dùng cách: for trên index; KHÔNG dùng hàm split(), len() 
s = input("Nhập vào một chuỗi: ")
kitu = 0
for char in s:
    kitu += 1
tu = 0
for i in range(kitu):
    if s[i] != " ":
        if i == 0 or s[i-1] == " ":
            tu += 1
print("Số từ trong chuỗi là:", tu)
```

```python
#W4A28 
#Nhập vào một chuỗi, hãy in từ đầu tiên trong chuỗi 
#Dùng cách: for trên index; KHÔNG dùng hàm split() 
s = input("Nhập vào một chuỗi: ")
kitu = 0
for char in s:
    kitu += 1
for i in range(kitu):
    if s[i] != " ":
        if i == 0 or s[i-1] == " " and s[i].isalpha():
            print("Từ đầu tiên trong chuỗi là:",s[i])
```

```python
#W4A29 
#Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành thường và thường thành hoa. 
n=input()                         
for i in range (len(n)): 
    b=ord(n[i])
    if 97<=b<=122:
        c=chr(b-32)
        print(f"{c}",end="")
    elif 65<=b<=90:
        d=chr(b+32)
        print(f"{d}",end="")
```

```python
#W4A29 
#Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành thường và thường thành hoa.
n=input()
kq='' 
for char in n: 
    if char.islower():
        kq+=char.upper()
    elif char.isupper():
        kq+=char.lower()
print(kq)
```

```python
#Bài 1: Nhập vào số nguyên n, in ra màn hình số gấp đôi của n

n=int(input("nhập vào số nguyên n"))
a=n*2
print(a)
```

```python
#Bài 2: Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. 
#Người ra xây một khu vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần diện tích còn để trồng cây xung quanh là bao nhiêu. 
#Biết pi = 3.14, lấy kết quả đến 2 chữ số phần thập phân. 

a,b=map(float,(input().split()))
if (a*b)-(3.14*a*a)>0:
    print("Diện tích còn để trồng cây xung quanh là" ,round((a*b)-(3.14*a*a),2))
else:
    print("Không còn diện tích trồng cây")
```

```python
#Bài 3: Viết chương trình nhập vào một “kí tự” c, 
#nếu nó là chữ hoa thì in ra chữ thường tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó. 
#(xem trước cấu trúc lựa chọn: if….else, ma) 

c=input()
if c.isupper():
    print(c.lower())
else:
    print(c.upper())
```

```python
#Bài 4: Viết chương trình nhập vào một “kí tự”,
#Kiểm tra xem c có phải là kí tự chữ cái không (hoa hoặc thường). 
#Nếu đúng thì in ra “{c} là kí tự alphabet”, nếu sai in ra “{c} không phải là kí tự alphabet"

c=input()
if c.isalpha():
    print(f'{c} là kí tự alphabet')
else:
    print(f'{c} không phải là kí tự alphabet')
```

```python
#Bài 5: Nhập vào chữ cái hoa A, in ra chữ cái thường liền trước chữ cái a thường tương ứng của A. Xét TH khi c = ‘A’ là đặc biệt.
c=input()
if c=="A":
    print("z")
else:
    print(chr(ord(c.lower())-1))
```

```python
#Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không, 
# nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, 
# nếu nó không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”.
a,b,c=map(float,(input().split()))
if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    S = (p * (p - a) * (p - b) * (p - c))**0.5
    print(f"Dien tich tam giac: {S:.1f}")
else:
    print("Khong phai 3 canh cua tam giac")
```

```python
#1 Kiểm tra số chẵn lẻ: Viết chương trình nhập một số nguyên, kiểm tra và in ra số đó là số chẵn hay số lẻ.
a=int(input())
if a%2==0:
    print("chẵn")
else:
    print("lẻ")
```

```python
#2  Kiểm tra số có tận cùng 5 không: 
# Viết chương trình nhập một số nguyên, kiểm tra số đó có tận cùng 5 không, nếu đúng in ra “True”, nếu sai in ra “False”.
a=int(input())
if a%5==0 and a%2!=0:
    print("True")
else:
    print("False")
```

```python
#3 Kiểm tra số có chia hết cho 3 và 5 không: 
# Viết chương trình nhập một số nguyên,kiểm tra số đó có chia hết cho 3 và 5 không, nếu đúng in ra “True”, nếu sai in ra “False”. 
a=int(input())
if a%3==0 and a%5==0:
    print("True")
else:
    print("False")
```

```python
#4 Kiểm tra tuổi đủ điều kiện bầu cử: 
# Viết chương trình nhập năm sinh của một người,  kiểm tra xem người đó có đủ 18 tuổi để đi bầu cử hay không.
a=int(input())
if 2026-a>=18:
    print("Đủ")
else:
    print("Chưa Đủ")
```

```python
#5 So sánh 2 số: Viết chương trình nhập vào hai số, in ra số lớn hơn hoặc thông báo nếu hai số bằng nhau
a,b=map(float,input().split())
if a>b:
    print(a)
elif b>a:
    print(b)
else:
    print(f"{a}={b}")
```

```python
#6 Kiểm tra ký tự nhập vào là chữ hay số: 
# Viết chương trình nhập một ký tự, kiểm tra xem ký tự đó là chữ cái hay số.
n=input()
if n.isupper() or n.islower():
    print("kí tự")
elif n.isdigit():
    print("số")
```

```python
#7 Bài tập 5: Tính điểm học lực 
# Nhập điểm (0-10), phân loại học lực theo quy tắc: >=8: Giỏi; >=6.5 và <8: Khá; >=5 và <6.5: Trung bình; <5: Yếu
a=float(input())
if a>=8:
    print("Giỏi")
elif 8>a>=6.5:
    print("Khá")
elif 6.5>a>=5:
    print("Trung Bình")
else:
    print("Yếu")
```

```python
#8 Kiểm tra năm nhuận: 
# Viết chương trình nhập năm, kiểm tra xem năm đó có phải năm nhuận hay không. 
# (Năm nhuận nếu chia hết cho 400 hoặc chia hết cho 4 nhưng không chia hết cho 100)
a=int(input())
if a%400==0 or a%4==0 and a%100!=0:
    print("Năm nhuận")
else:
    print("Năm không nhuận")
```

```python
#9 Chuyển đổi số thành chữ số: 
# Nhập một số nguyên từ 0 đến 9, in ra tên chữ số tương ứng (0 -> "Không", 1 -> "Một", ...) – Xem trước match …case 
def cstu(n):
    match n:
        case 0:
            return "Không"
        case 1:
            return "Một"
        case 2:
            return "Hai"
        case 3:
            return "Ba"
        case 4:
            return "Bốn"
        case 5:
            return "Năm"
        case 6:
            return "Sáu"
        case 7:
            return "Bảy"
        case 8:
            return "Tám"
        case 9:
            return "Chín"
n=int(input())
print(cstu(n))
        
```
